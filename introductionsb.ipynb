```python
from google.colab import drive
from IPython.display import HTML
from base64 import b64encode
import shutil

def play_video_from_drive():
    drive.mount('/content/drive')


    source_path = '/content/drive/My Drive/projectvideo/simone.mp4'  # Path in Google Drive
    destination_path = '/content/simone.mp4'  # Local path in Colab

    shutil.copy(source_path, destination_path)
    print("File copied to local directory.")

    with open(destination_path, "rb") as video_file:
        video_encoded = b64encode(video_file.read()).decode('ascii')

    return HTML(f"""
    <video width="640" height="480" controls autoplay loop>
        <source src="data:video/mp4;base64,{video_encoded}" type="video/mp4">
    </video>
    """)

play_video_from_drive()
```
